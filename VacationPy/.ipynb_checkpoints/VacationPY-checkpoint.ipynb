```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

```python
# Load the csv exported in Part I to a DataFrame

city_df = pd.read_csv("../WeatherPy/output_data/weatherpy_data.csv")

city_df.head()
```

```python
# city_df["humidity"].value_counts()
```

```python
# Configure gmaps.
gmaps.configure(api_key=g_key)

# Use the Lat and Lng as locations and Humidity as the weight.
city_locations = city_df[["lat", "lng"]]

# Add Heatmap layer to map.

# Making fig bigger
figure_layout = {
    'width': '1000px',
    'height': '900px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}

# Creating g.maps figure with desired layout
fig = gmaps.figure(layout=figure_layout)

# Creating heatmap layer
heat_layer = gmaps.heatmap_layer(city_locations, weights = city_df["humidity"], dissipating = False, 

                                 max_intensity = 100, point_radius = 3)

# Creaing hotel markers / only after its been ran
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)

# Adding heat map layer
fig.add_layer(heat_layer)

# Adding hotel markers / only after its been ran
fig.add_layer(markers)


# Showing the map
fig
```

```python
# Narrow down the cities to fit weather conditions.

# * Narrow down the DataFrame to find your ideal weather condition. For example:

#   * A max temperature lower than 80 degrees but higher than 70.

#   * Wind speed less than 10 mph.

#   * Zero cloudiness.

#   * Drop any rows that don't contain all three conditions. You want to be sure the weather is ideal.

hotel_df = city_df.loc[(city_df["cloudiness"] == 0) & (city_df["wind speed"] < 10) & (city_df["max temp"] > 70) & (city_df["max temp"] <80)].copy()

hotel_df
```

```python
# Creating an empty column called hotel name
hotel_df["hotel name"] = ""

hotel_df.head()
```

```python
# Setting params dictionary for api call
params = {"radius" : 5000,
          "keyword" : "hotel",
          "key" : g_key
}

# Nearby search link for setting up call
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# Looping thru dataframe using iterrows to grab desired data and add it to empty column hotel name
for index, row in hotel_df.iterrows():
    # Grabbing the lat and lng from the dataframe for later...
    lat = row["lat"]
    lng = row["lng"]
    
    # Adding lat and lon to params dictionary, needed to complete the api call
    params["location"] = f"{lat}, {lng}"
    
    # Getting data and storing as a json using base url and params dictionary
    hotel_names = requests.get(base_url, params = params).json()
    
    # Indexing into row and adding the name of the hotel
    hotel_df.loc[index, "hotel name"] = hotel_names["results"][0]["name"]
```
